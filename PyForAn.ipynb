```python
import pandas as pd
df = pd.read_csv("kc-house-data (1).csv", sep = ',', encoding = 'windows-1251')
df.head(5)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.plot(df["price"],)
plt.title('Распределение по стоимости')
plt.xlabel('Количество')
plt.ylabel('Стоимость')
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["sqft_living"],)
plt.title('Распределение жилой площади')
plt.xlabel('Количество')
plt.ylabel('Жилая площадь')
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["yr_built"],)
plt.title('Распределение года постройки')
plt.xlabel('год постройки')
plt.ylabel('Количество')
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["waterfront"],)
plt.title('Распределение по виду на набережную')
plt.xlabel('вид на набережную')
plt.ylabel('Количество')
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["floors"],)
plt.title('Распределение по этажам')
plt.xlabel('Количество этажей')
plt.ylabel('Количество')
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["grade"],)
plt.title('Распределение по состоянию')
plt.xlabel('Состояние дома в баллах')
plt.ylabel('Количество')
```

```python
plt.figure(figsize=(6,4))
x=df['price']
y=df['sqft_living']
plt.plot(x,y)
```

Из графика видно , что в общем с увеличением жилой площади возрастает стоимость
дома

```python
plt.figure(figsize=(6,4))
x=df['price']
y=df['bedrooms']
plt.plot(x,y)
```

Из графика видно , что стоимость дома растет с увеличением числа спален

```python
plt.figure(figsize=(6,4))
x=df['price']
y=df['yr_built']
plt.plot(x,y)
```

Из графика видно , что стоимость дома хоть и зависит от года постройки ,но невыраженно
